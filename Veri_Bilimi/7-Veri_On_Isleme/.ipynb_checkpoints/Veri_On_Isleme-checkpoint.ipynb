# Aykırı Gözlem Analizi <a class="anchor" id="aykiri-gozlem-analizi"></a>

## Aykırı Gözlem <a class="anchor" id="aykiri-gozlem"></a>

Veride genel eğilimin oldukça dışına çıkan ya da diğer gözlemlerden oldukça farklı olan gözlemlere **aykırı gözlem** denir.

Aykırılığı ifade eden nümerik değere **aykırı değer** denir.

Aykırı değeri barındıran gözlem birimine **aykırı gözlem** denir.

**Not:** Arada çok fark olmamakla beraber bilinmesinde fayda vardır. Biri diğerinin yerine kullanılabilir.

**EN ÖNEMLİ PROBLEMLERDEN BİRİSİDİR.**

**Aykırı Gözlem Neye Sebep Olur?**  


Grafiği inceleyecek olursak normalde X-Y arasındaki ilişki, doğruya bakıldığında negatif yönlü iken aykırı gözlemler yüzünden doğrunun eğimi pozitif yönlü olmuş. Yani aykırı gözlemler değişkenler arasındaki ilişkinin **genel yönünü** bozmuştur. Bu yüzden aykırı gözlemler bu gibi sorunlara sebep olabilmektedir.

**1. Sektör Bilgisi** 

Aykırı gözlem tespiti için uygulanacak formülleri kullanmak yerine sektör kapsamında uzman kişi yardımıyla veya kendimiz hangi gözlemlerin aykırı olduğunu bulabiliriz.

**2. Standart Sapma Yaklaşımı**

**Not:** 1,2 ve 3 standart sapma değeri örnekten örneğe değişebilir.

**Örnek**  


**3. Z-Skoru Yaklaşımı**

**4. Boxplot(Interquartile Range - IQR) Yöntemi**

Bu yöntemlerden genellikle 1. ve 4. yöntemleri kullanmaya çalışacağız. Sektör bilgimiz varsa 1. yöntemi, yoksa ve tek değişkenli bir aykırı gözlem incelemesi yapacaksak 4. yöntemi kullanacağız.

## Aykırı Değerleri Yakalamak <a class="anchor" id="aykiri-degerleri-yakalamak"></a>

Tek değişkenli olarak ele alınacaktır.

```python
df = sns.load_dataset("diamonds")
df = df.select_dtypes(include = ["float64", "int64"]) #Sayısal değişkenler seçildi.
df = df.dropna() #Eksik değerler silindi. (Aykırı değer tespitinin kolay yapılması için silindi. İleriki derste detaylı ele alınacaktır.)
df.head()
```

Sayısal değişkenler geldi. Burada **table** değişkeninin aykırı değerlerini tespit edeceğiz.

```python
df_table = df["table"]
```

```python
df_table.head()
```

Boxplot yöntemini kullanarak aykırı değerleri tespit edelim. Öncelikle grafiğini çizdirerek aykırı değerleri gözlemleyelim:

```python
sns.boxplot(x = df_table);
```

Görüldüğü üzere grafikteki noktalar bizim aykırı değerlerimizdir. Şimdi bu yöntemden bir eşik değer hesaplayalım:

```python
Q1 = df_table.quantile(0.25) #Çeyreklik hesaplama fonksiyonu. (Varsayılan: 0.5)
Q3 = df_table.quantile(0.75)
IQR = Q3 - Q1
```

```python
Q1
```

```python
Q3
```

```python
IQR
```

```python
alt_sinir = Q1 - 1.5 * IQR
ust_sinir = Q3 + 1.5 * IQR
#Konu anlatımdan farklı olarak bu formül de uygulanabilir.
```

```python
alt_sinir
```

```python
ust_sinir
```

Alt sınırları ve üst sınırları belirledik. Şimdi bunu `df_table` üzerinden bu şartları koyarak değerleri tespit edelim:

```python
(df_table < alt_sinir) | (df_table > ust_sinir)
```

Örnek olması açısından alt sınırda olanları alalım:

```python
aykiri_tf_alt = (df_table < alt_sinir)
```

```python
aykiri_tf_alt.head()
```

Görüldüğü üzere True-False şeklinde değerler döndürdü. Şimdi **True** olanların kendi değerlerine erişelim:

```python
df_table[aykiri_tf_alt]
```

Görüldüğü üzere `[]` içerisine yazarak **True** değerlerine eriştik. Şimdi bunların indexlerini çekelim:

```python
df_table[aykiri_tf_alt].index
```

Bu indexlere erişmemizin sebebi bu aykırı değerlerin indexlerini kullanarak silme işlemi, ortalamaya çekme işlemi ya da baskılama işlemini gerçekleştirebiliriz.

## Çok Değişkenli Aykırı Gözlem Analizi <a class="anchor" id="cok-degiskenli-aykiri-analizi"></a>

### Local Outlier Factor

Gözlemleri bulundukları konumda yoğunluk tabanlı skorlayarak buna göre aykırı değer olabilecek değerleri tanımlayabilmemize imkan sağlar.

Bir noktanın local yoğunluğu bu noktanın komşuları ile karşılaştırılıyor. Eğer bir nokta, komşularının yoğunluğundan anlamlı şekilde düşük ise **"bu nokta komşularından daha seyrek bir bölgede bulunuyordur"** yorumu yapılabiliyor. Dolayısıyla burada bir komşuluk yapısı söz konusudur. **"Bir değerin çevresi yoğun değilse demek ki bu değer aykırı değerdir"** şeklinde değerlendiriliyor.

Bu yöntemin ortaya çıkış sebebi tek değişken için ilgili değişkenler aykırı değer olmazken çok değişkenli şeklinde bakıldığında aykırı değer olabilir. En çok kullanılan yöntemlerden birisidir.

**Örnek:**  


**Evlilik Sayısı** değişkenine tekil olarak bakıldığında 3 değeri aykırı bir değer olmazken **Yaş** değişkeniyle beraber bakıldığında 17 yaşındaki birisinin 3 evlilik yapması aykırı bir durumdur. Gerçekte olsa bile genel yapının dışında olduğundan aykırı gözlem olarak algılanır.

Resim üzerinden bunu inceleyelim:

En yakın üç komşuya bakılacak olursa A noktasının diğerlerine göre komşusu daha uzak olduğundan aykırı değer olarak algılanır.

```python
diamonds = sns.load_dataset("diamonds")
diamonds = diamonds.select_dtypes(include = ["float64", "int64"])
df = diamonds.copy()
df = df.dropna()
df.head()
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
clf = LocalOutlierFactor(n_neighbors=20, contamination=0.1)

#n_neighbors: Komşu sayısı
#contamination: Yoğunluk
```

```python
clf.fit_predict(df)
```

```python
df_scores = clf.negative_outlier_factor_
```

```python
df_scores[0:10]
```

Her bir değer için skorları hesapladı. Bunları sıralayacak olursak:

```python
np.sort(df_scores)[0:20]
```

Görüldüğü üzere skorlar sıralandı. Bundan sonra belirli kriterlere (proje hassaslığı, gözlem çokluğu vs.) ve yönteme göre bir eşik değer belirlememiz gerekiyor fakat örnek yaptığımızdan dolayı gözlem yaparak bulabiliriz.  
Örneğin 13. skoru eşik değer olarak belirleyelim:

```python
esik_deger = np.sort(df_scores)[13]
esik_deger
```

Sonra eşik değerden büyük olanları alalım. Böylelikle aykırı değerleri almamış oluruz.

```python
aykiri_tf = df_scores > esik_deger
```

```python
aykiri_tf
```

True şeklinde geldi. Şimdi bu değerleri getirelim:

**Silme Yöntemi**

```python
yeni_df = df[aykiri_tf]
```

```python
yeni_df
```

Aykırı değerleri getirelim:

```python
df[df_scores < esik_deger]
```

Önceki derste tek değişkenli olarak **"table"** değişkenini inceledeğimizde ona kıyasla çok daha az aykırı gözlemler geldi.

## Baskılama <a class="anchor" id="baskilama"></a>

Eşik değere ilişkin gözlemi getirelim:

```python
df[df_scores == esik_deger]
```

```python
baski_deger = df[df_scores == esik_deger]
```

```python
aykirilar = df[~aykiri_tf]
#"aykiri_tf = df_scores > esik_deger" idi. Bunu sağlamayanları (eşik değer de dahil) getir demektir.
#Çünkü bunun zıt koşulu: df_scores <= esik_deger şeklindedir.
#Yani burada eşik değerin kendisi de alındı.

aykirilar
```

Eşik değeri ve aykırıları değişkenlere kaydettik. Bu aykırıları baskı değerine (eşik değeri) eşitlememiz için bunların indexlerini kaldırıp değerleri numpy array'ine çevirmemiz gerekiyor. Aksi takdirde index sorunlarına yol açabiliyor.

```python
res = aykirilar.to_records(index = False)
res
```

Şimdi eşitlememizi yapalım:

```python
res[:] = baski_deger.to_records(index = False)

#Kod pratikliği açısından baskı değerindeki gözlemin indexini direkt kaldırıp eşitledik.
```

```python
res
```

Görüldüğü üzere aykırı gözlemler eşik değerdeki gözlemle aynı olmuş oldu.  
Şimdi ise bu **res** değişkenini DataFrame'e çevirip aykırı gözlemlerin olduğu DataFrame'e eşitlememiz gerekiyor.

```python
df[~aykiri_tf]
```

Yani değişkenimiz bu DataFrame'e eşitlenmelidir.

```python
df[~aykiri_tf] = pd.DataFrame(res, index = df[~aykiri_tf].index)
```

```python
df[~aykiri_tf]
```

# Eksik Gözlem Analizi <a class="anchor" id="eksik-gozlem-analizi"></a>

## Eksik Gözlem <a class="anchor" id="eksik-gozlem"></a>

Eksik veri, incelenen veri setindeki gözlemlerin eksiklik olması durumunu ifade etmektedir.

Böyle bir durumla karşılaşıldığında eksik veriyi ya silebiliriz, ya da doldurabiliriz.

**Eksik Veriyi Direkt Silmenin Zararları**  


**Eksik Verilerde Bilinmesi Gerekenler**  


Tek değişkenli olarak bakıldığında eksik veri olarak gözükse de başka bir değişkenle bakıldığında aslında eksik olmasının mantıklı olduğunu görürüz. **Eksik veri incelemesinde bu çok dikkat edilmelidir.** Yani açıklamada dendiği gibi bu **eksik veriler başka değişkenlerle bağımlı olmamalı, rastgele oluşmalı.** Böyle bir durumda veriyi silmek mantıksız olacaktır.

Örneğin bu örnekte kişinin kredi kartı var fakat harcaması **NA** olarak gözüküyor. Bu durum o verinin eksik veya boş olduğu anlamına gelmez, kişinin o ay hiç harcama yapmadığı anlamına da gelebilir. Bu yüzden o veri 0 olarak kabul edilmelidir. Yani **NA** her zaman eksiklik değildir.

100 değişkenli bir veri seti için sadece 1 tane eksik verinin olması o satırı komple silmemiz gerektiği anlamına gelmez. Eğer bu yapılırsa çok büyük oranda bilgi kaybına sebep olur. Bu durumda doldurma yapmak mantıklı olacaktır.

**Eksik Veri Türleri**  


Bu türlerin bilinmesi çok önemlidir. Çünkü bu türlere göre gerekli silme veya doldurma yöntemleri uygulanır. En çok sevilen tür **tümüyle rastlantısal kayıptır.** Çünkü bu durumda eksik veri gönül rahatlığıyla silinebilir veya ortalama ile doldurulabilir.

**Eksik Veri Rassallığının Testi**  


En çok kullanılan testler **görsel teknikler** ve **Little'ın MCAR testidir.** Bu eğitimde daha çok görsel teknikler ile test edeceğiz.

**"Atama fikri hem çekici hem de tehlikelidir."** -R.J.A. Little & D.B. Rubin

**Eksik Veri Problemi Nasıl Giderilir?**

## Eksik Veri Hızlı Çözüm <a id="eksik-veri-hizli-cozum"></a>

Eksik veri analizi uzun bir analiz olduğundan bu analize ihtiyaç duymayan veri setleri ve hızlı bir çözüm isteyenler için bu çözüm yardımcı olacaktır. Fakat bu şekilde yapmanın sorunlara yol açacağını söylemekte fayda var.

```python
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])
df = pd.DataFrame({
    "V1" : V1,
    "V2" : V2,
    "V3" : V3
})
df
```

Eksik değerlerin değişken bazında toplam sayısına bakalım:

```python
df.isnull().sum()
```

Eksik değer olmayanların sayısına bakalım:

```python
df.notnull().sum()
```

Veri setindeki **toplam eksik değerlere** bakalım:

```python
df.isnull().sum().sum()
```

```python
df.isnull()
```

En az 1 tane **NaN** olan satırları getirelim. Bunun için köşeli parantez içerisine `any()` fonksiyonunu da kullanmamız gerekiyor:

```python
df[df.isnull().any(axis = 1)] #axis = 1, sütun için bak demektir.
```

Şimdi hiç **NaN** değeri olmayan satırları getirelim:

```python
df[df.notnull().all(axis = 1)]
```

Başka bir yöntemle yapmak istersek:

```python
df[df["V1"].notnull() & df["V2"].notnull() & df["V3"].notnull()] #all() fonksiyonunun kod hali
```

Şimdi bu problemi giderelim:

**Eksik Gözlemlerin Direkt Silinmesi**

```python
df.dropna()
```

Görüldüğü üzere `dropna()` fonksiyonu bir tane bile **NaN** değeri olan bütün satırları sildi. Fakat bunu kalıcı olarak yapmadı. `df` değişkenimiz hala normal.

```python
df
```

Kalıcı silme yapmak için:

```python
df.dropna(inplace = True)
```

```python
df
```

Görüldüğü üzere kalıcı olarak silindi.

Başka örnek için df'yi tekrar normal hale getirelim:

```python
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])
df = pd.DataFrame({
    "V1" : V1,
    "V2" : V2,
    "V3" : V3
})
df
```

**Basit Değer Atama**

Eksik değerleri her bir değişkenin ortalamasını atayalım:

```python
df["V1"]
```

**V1** değişkeninin eksik değerlerine ortalama atayalım:

```python
df["V1"].mean()
```

```python
df["V1"].fillna(df["V1"].mean())
```

**V2** değişkeninin eksik değerlerine 0 atayalım:

```python
df["V2"].fillna(0)
```

Görüldüğü üzere eksik değerler dolduruldu.  
Birden fazla değişken için bu işlemleri tek tek yapmak zor olacaktır. Bunu otomatik gerçekleştirmek için `apply()` ve `lambda` mimarilerini kullanmalıyız. Çünkü `apply()` fonksiyonu DataFrame'in tüm sütunlardaki değişkenlere tek seferde değişiklik yapabilmemizi sağlar.

df'yi yeniden eski haline getirelim:

```python
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])
df = pd.DataFrame({
    "V1" : V1,
    "V2" : V2,
    "V3" : V3
})
df
```

```python
df.apply(lambda x: x.fillna(x.mean()), axis = 0)
```

Görüldüğü üzere `lambda` fonksiyonunu kullanarak eksik değerleri ortalama ile doldurmuş olduk.

Başka bir yöntemle inceleyelim:

## Kategorik Değişken Kırılımında Değer Atama <a id="kat-deg-kirilimi-deger-atama"></a>

Örneğin bir şirketin maaşları ile ilgili bir veri setinde maaş değişkeninde bazı değerler boş olmuş olsun. Bunu direkt ortalama ile atamak bazı durumlarda mantıklı olmayacaktır. Çünkü örneğin maaşı boş olan kişi alt skalada veya üst skalada bir departmanda ise genel kitleye göre ortalama atamak doğru değildir, **çalıştığı departmanın maaş ortalamasına göre boş değer doldurulmalıdır.** Bu yüzden böyle durumlarda diğer kategorik değişkenleri inceleyip ona göre atama işlemi yapmak daha mantıklıdır.

Örnek üzerinden bunu inceleyelim:

```python
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])
V4 = np.array(["IT","IT","IK","IK","IK","IK","IK","IT","IT"])
df = pd.DataFrame({
    "maas" : V1,
    "V2" : V2,
    "V3" : V3,
    "departman" : V4
})
df
```

```python
df.groupby("departman")["maas"].mean()
```

Şimdi maaş değişkenindeki boş değerleri departman bazında ortalamalarına göre dolduralım:

```python
df["maas"].fillna(df.groupby("departman")["maas"].transform("mean"))
```

Görüldüğü üzere departman bazında eksik değerler dolmuş oldu. `transform()` fonksiyonunu yazmamızın sebebi eksik değerleri ortalama ile "dönüştürme" yapmamız gerektiği içindir. Zaten `fillna()` içerisine `df.groupby("departman")["maas"].mean()` şeklinde yazınca bir değişiklik olmayacaktır. Deneyecek olursak:

```python
df["maas"].fillna(df.groupby("departman")["maas"].mean())
```

Görüldüğü üzere bir değişiklik olmadı. Çünkü `df.groupby("departman")["maas"].mean()` ifadesi bir dönüştürme işlemi değil, sadece gösterme kodudur. Bu yüzden `transform()` fonksiyonu kullanılmalıdır.

## Kategorik Değişkenlerde Değer Atama <a id="kat-deg-deger-atama"></a>

Kategorik değişkenler için değer atamanın en mantıklı ve verimli yolu mod işlemidir. Yani değişkende en sık tekrar eden kategori ne ise o kategori ile doldurulur.

```python
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V4 = np.array(["IT",np.nan,"IK","IK","IK","IK","IK","IT","IT"], dtype=object)
df = pd.DataFrame({
    "maas" : V1,
    "departman" : V4
})
df
```

Departman değişkeninin modunu alalım:

```python
df["departman"].mode()
```

0 indexini al dersek:

```python
df["departman"].mode()[0]
```

Bizlere "IK" sonucunu döndürür. Şimdi bu kodumuzu `fillna()` fonksiyonuna yazalım:

```python
df["departman"].fillna(df["departman"].mode()[0])
```

Görüldüğü üzere eksik değer değişkenin modu olan "IK" ile dolmuş oldu. Kalıcı yapmak için ise `inplace` argümanı kullanılabilir.

Modunu almak yerine bir önceki veya bir sonraki kategori ile doldurmak isteyebiliriz. Şimdi bunu yapalım:

Sonraki kategoriyle doldurmak için:

```python
df["departman"].fillna(method = "bfill")
```

Önceki değerle doldurmak için:

```python
df["departman"].fillna(method = "ffill")
```

Görüldüğü üzere `method` argümanını kullanarak önceki ve sonraki kategoriyle doldurmuş olduk.

## Değişken Standardizasyonu (Veri Standardizasyonu) <a id="degisken-stand"></a>

Değişken dönüşümünden farklı olarak bu yöntem, değişkenin kendi içinde varyans/bilgi yapısı bozulmaz. Fakat belirli bir standarda oturtulur.  
Değişken dönüşümünü ise değişkenin taşıdığı bilgiyle kalamamasıdır. (örn: kadın-erkek yerine 0-1 yazmak.) Aslında özünde bilgi değişmez fakat yapısı değişir.  
Bu dönüştürme işlemini yapmamızın sebebi veriyi göndereceğimiz fonksiyonların genelde bu şekilde istemesinden dolayıdır.

```python
V1 = np.array([1,3,6,5,7])
V2 = np.array([7,7,5,8,12])
V3 = np.array([6,12,5,6,14])
df = pd.DataFrame({
    "V1" : V1,
    "V2" : V2,
    "V3" : V3,
})
df = df.astype(float)
df
```

**Standardizasyon**

Verileri standart normal dağılıma dönüştürme işlemidir. (0 ortalamalı 1 varyanslı)

```python
from sklearn import preprocessing
```

```python
preprocessing.scale(df)
```

**Not:** Makine öğrenmesi algoritmaları _genelde_ farklı ölçek türlerdeki verileri çok sevmemektedir. Bu yüzden verileri standartlaştırmak gerekir. Fakat bazı makine öğrenmesi algoritmaları için bu gerekmemektedir.

```python
df
```

Orijinal yapısı bozulmadı. Dönüştürme fonksiyonları genelde orijinal yapıyı bozmaz. (`copy = True` veya `inplace = False`)

```python
df.mean()
```

**Normalizasyon**

Değerleri 0 ile 1 arasına dönüştürür.

```python
preprocessing.normalize(df)
```

**Max - Min Dönüşümü**

Bizim istediğimiz 2 aralığa dönüştürmek için kullanılır.

```python
scaler = preprocessing.MinMaxScaler(feature_range = (10,20))

#MinMaxScaler fonksiyonundan bir örnek oluşturduk. Bu nesne artık belirli özellikleri olan bir dönüştürücü nesnesidir.
#Şu anki özelliği, verileri 10-20 arasına çekmektir.
```

```python
scaler.fit_transform(df)

#Bu nesneyle beraber fit_transform fonksiyonunu kullanarak dönüştürme işlemini gerçekleştirdik.
#Yani belirli özellikte bir nesne oluşturduk ve bunu artık fit_transform fonksiyonuyla df'e uygula dedik.
```

## One Hot Dönüşümü Ve Dummy Değişken Tuzağı <a id="one-hot-ve-dummy-deg"></a>

Bu dönüşüm yapılırken bir dummy (kukla) değişken tuzağı karşımıza çıkıyor. Bu durumları bir ele alalım:

```python
df.head()
```

```python
df_one_hot = pd.get_dummies(df, columns=["sex"], prefix=["sex"])

#prefix: ön isimlendirme
```

```python
df_one_hot.head()
```

Veriler dikkatle incelendiğinde orijinal cinsiyet değişkenin gittiğini ve en sağda cinsiyet kategorisinin sayısı kadar değişken oluştuğunu görüyoruz. 1 olan değerler aslında kendisini ifade etmektedir (yani male=1 ise bu kişi erkektir, kadın 0'dır; female=1 ise bu kişi kadındır, erkek 0'dır). Fakat burada ikisi de aynı bilgiyi verdiği için birisi dummy değişkendir. Bu dönüşüm sonrasında oluşturulan yeni değişkenler birbiri üzerinden oluşturabiliyorsa, ya da bir değişkeni ifade eden başka bir değişken varsa buna **dummy değişken** denir. Bir değişkende kaç kategori varsa, 1 eksiği kadar dummy değişken var demektir.

Şimdi birden fazla kategorisi olan `day` değişkeni için bunu yapalım:

```python
pd.get_dummies(df, columns=["day"], prefix=["day"])
```

Bu veri incelendiğinde değişkenin kategori sayısı kadar yeni değişkenler oluşturdu.

One-Hot dönüşümünün bize faydası ise
1. Kategorilerin etkileri, makine öğrenmesi algoritmasına hissettirilebilmektedir. Yani örneğin bir değişkenin herhangi bir kategorisinin etkisi büyükse, one-hot dönüşümü ile o kategori 1 olup diğerleri 0 olacağından dolayı bu fark anlaşılacak ve makine öğrenmesinin performansı artacaktır.
2. Oluşan değişkenler birbiri üzerinden oluşamayacağından ve bir değişkeni ifade eden başka değişkenler olamayacağından dolayı çok kategorili değişkenlerde dummy değişken olmayacaktır. Yani örneğin yeni oluşan tek bir değişken incelendiğinde oluşan diğer değişkenlerle aynı bilgiyi vermemektedir, kendine özgü bilgi vermektedir. Fakat 2 kategorili kadın-erkek değişkeni için aynı bilgileri verdiklerinden dolayı birisi dummy değişkendir. Bu yüzden çok kategorili durumlarda da one-hot dönüşümü bizim için faydalıdır.
